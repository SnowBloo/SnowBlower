```python
from skimage.io import imread, imshow
from skimage import transform
import matplotlib.pyplot as plt
import numpy as np

import cv2
```

```python
chess = imread('test_image.jpg')
imshow(chess)
```

```python
#source coordinates
src = np.array([1560, 1582, 
                620, 1620,
                666, 1908,
                1998, 1808,]).reshape((4, 2))#destination coordinates
dst = np.array([500, 100, 
                100, 100,
                100, 500,
                500, 500,]).reshape((4, 2))#using skimages transform module where projective is our desired parameter
# tform = transform.estimate_transform('projective', src, dst)
# tf_img = transform.warp(chess, tform.inverse)#plotting the transformed image

h, status = cv2.findHomography(src, dst)
tf_img = cv2.warpPerspective(chess, h, (400, 400))
fig, ax = plt.subplots()
ax.imshow(tf_img)
_ = ax.set_title('projective transformation')
```
